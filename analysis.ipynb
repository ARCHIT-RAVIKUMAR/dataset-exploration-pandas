# Business Problem

Netflix aims to improve its content strategy by understanding the characteristics of its existing catalog. One possible business problem is predicting whether newly acquired content should be categorized as a **Movie** or a **TV Show** based on its attributes.

This can support:
- Better content acquisition decisions
- Improved recommendation systems
- Efficient catalog management
- Production planning

# Machine Learning Problem Framing

**Problem Type:** Classification

The objective is to classify content into one of two categories: **Movie** or **TV Show**.

### Target Variable

- `type`

### Key Features

- release_year
- duration
- rating
- country
- listed_in
- director

Since the target variable contains predefined categories, this is a Classification problem.

# Key Observations

1. Movies make up a significantly larger portion of the Netflix catalog than TV Shows.

2. Most content in the dataset was released after 2015, reflecting Netflix's rapid expansion in recent years.

3. Columns such as `director`, `cast`, and `country` contain missing values, indicating that data preprocessing will be required before building a machine learning model.
